이 코드는 `numpy`를 사용하여 이산 정현파 데이터를 생성하고 `matplotlib`을 사용하여 그래프를 그립니다.

# 새 섹션

```python
import numpy as np
import matplotlib.pyplot as plt

# 주파수 (Hz)
f = 10
# 샘플링 주파수 (Hz)
fs = 100
# 시간 주기 (초)
T = 1/f
# 샘플링 주기 (초)
ts = 1/fs

# 시간 벡터 생성
t = np.arange(0, T, ts)

# 정현파 생성
signal = np.sin(2 * np.pi * f * t)

# 그래프 그리기
plt.figure(figsize=(10, 4))
plt.stem(t, signal)
plt.title('Discrete Sine Wave')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
